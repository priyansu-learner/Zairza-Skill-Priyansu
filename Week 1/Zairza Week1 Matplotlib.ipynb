```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
x = np.arange(0,10,0.1)
y = np.sin(x)

plt.title("A SIne Wave")
plt.xlabel("x")
plt.ylabel("Sin(x)")

plt.plot(x,y)

plt.show()
```

```python
x = np.arange(0,10,0.4)
y = np.sin(x)

plt.title("A SIne Wave")
plt.xlabel("x")
plt.ylabel("Sin(x)")

plt.scatter(x,y,color="black")

plt.show()
```

```python
x = np.arange(0,10,0.4)
y = np.sin(x)

plt.title("A SIne Wave")
plt.xlabel("x")
plt.ylabel("Sin(x)")

plt.plot(x,y,marker='o',color="black")

plt.show()
```

```python
x = np.arange(0,10,0.1)
y = np.sin(x)
z = np.cos(x)

plt.title("A SIne Wave")
plt.xlabel("x")
plt.ylabel("Sin(x)")

plt.plot(x,y,x,z)

plt.show()
```

```python
a = np.arange(0,11)

plt.plot(a)
plt.plot(1.1*a+1,ls = "dashed")
plt.plot(1.2*a+2,ls = "dotted")
plt.plot(1.3*a+3,ls = "dashdot")
plt.plot(1.4*a+4)
plt.plot(1.5*a+5,ls = "dashed")
plt.plot(1.6*a+6,ls = "dotted")
plt.plot(1.7*a+7,ls = "dashdot")

plt.show()
```

```python
b = np.arange(0,11)

plt.plot(b,color="red")
plt.plot(1.1*b+1,ls = "dashed",color="green")
plt.plot(1.2*b+2,ls = "dotted",color="blue")
plt.plot(1.3*b+3,ls = "dashdot",color="black")

plt.show()
```
